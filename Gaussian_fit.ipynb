```python
data = np.array([63, 59, 74, 78, 70, 74, 75, 82, 68, 69,\
        76, 62, 72, 88, 65, 81, 79, 77, 66, 76,\
        86, 72, 79, 77, 60, 70, 65, 69, 73, 77,\
        72, 79, 65, 66, 70, 74, 84, 76, 80, 69])
```

```python
N = np.size(data)
mean = np.sum(data)/N
rms = np.sqrt( np.sum( (data-mean)**2)/(N-1) )
print(mean, rms)
```

```python
x = np.arange(59, 89, 1)
gaussian = 1/(np.sqrt(2*np.pi)*rms )*np.exp( -(x - mean)**2 /(2*rms**2) )



plt.figure(figsize = (12, 5))
plt.plot(x, gaussian, 'r')
plt.xlabel('Random Histogram')
plt.ylabel('Frequency',fontsize=15)
plt.hist(data, 10, normed = True, ec='black')#density = True
#plt.hist(data, 10, normed = True, density = True, ec='black')#
plt.legend(loc='upper right',fontsize=30)
plt.xticks(fontsize = 15) 
plt.yticks(fontsize = 15) 
plt.grid()
```
