$$
f(x) = 1\textrm{ if } 4n < x < 4n + 2\\
f(x) = 0\textrm{ if } 4n + 2 < x < 4n + 4
$$

```python
x = np.arange(-10, 10.01, 0.01)
N = len(x)
y = np.zeros(N) + 1

p = 0
while p < N:
    y[p*400: 200 + p*400] = -1
    p += 1

plt.figure(figsize = (11, 4))
plt.plot(x, y, Linewidth = 5)
plt.grid(True)
plt.show()
```

$$
y(t) = \frac{4}{\pi}\sum_{k= 1}^{\infty} \frac{\sin{2\pi\left( 2k - 1 \right)ft}}{2k-1}= \frac{4}{\pi}\left( \sin{\omega t} + 
\frac{1}{3}\sin{3\omega t} + \frac{1}{5}\sin{5\omega t}\dots \right)
$$

```python
n = 15
omega = 2*np.pi/4
j = 1
#fourier = 4*np.sin(j*omega*x)/(j*np.pi)
fourier = np.zeros(len(x))



for j in range(1, n, 2):
    #j = 1
    print('wave number ', j)
    lastWave = 4*np.sin(j*omega*x)/(j*np.pi)
    fourier = fourier + lastWave

    plt.figure(figsize = (11, 4))
    plt.plot(x, y, Linewidth = 4, color = 'blue')
    plt.plot(x, lastWave, Linewidth = 2, color = 'red')
    plt.plot(x, fourier, Linewidth = 2, color = 'k')
    plt.grid()
    plt.show()
    
```
